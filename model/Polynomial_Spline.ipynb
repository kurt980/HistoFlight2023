```python
import pandas as pd
import numpy as numpy
from datetime import date
```

```python
flights_df = pd.read_csv('C:/Users/KurtJi/OneDrive - University of Illinois - Urbana/Desktop/Personal Projects/HistoFlight2023/old_data/2022.10.26/flight.csv')
tickets_df = pd.read_csv('C:/Users/KurtJi/OneDrive - University of Illinois - Urbana/Desktop/Personal Projects/HistoFlight2023/old_data/2022.10.26/ticket.csv')
```

```python
# Join the two data frames based on the 'flight_id column
df = pd.merge(flights_df, tickets_df, on='flight_id', how='outer')
```

```python
df
```

```python
# introduce day-of-week variable
# convert the date column to datetime
df0 = df.dropna(axis=0)
df0['departure_date'] = pd.to_datetime(df0['departure_date'])
df0['purchase_date'] = pd.to_datetime(df0['purchase_date'])

# get the day of the week
df0['departure_day_of_week'] = df0['departure_date'].dt.day_name()
df0['departure_day_of_month'] = df0['departure_date'].dt.month_name()
df0['purchase_day_of_week'] = df0['purchase_date'].dt.day_name()
df0['purchase_day_of_month'] = df0['purchase_date'].dt.month_name()
```

```python
df0 = df0.drop(columns=['flight_number','flight_id', 'ticket_id', 'departure_date', 'arrival_date', 'purchase_date'],axis=1)
```

### Feature Engineering

```python
# Dummy variables
df0 = pd.get_dummies(df0, columns=['airline_code','departure_airport', 'arrival_airport', 'class', 'departure_day_of_week', 'departure_day_of_month', 'purchase_day_of_week', 'purchase_day_of_month'])
```

```python
# Date and time objects
# Convert the time of day data into a numeric format
df0['departure_time'] = pd.to_datetime(df0['departure_time']).dt.minute + \
                      pd.to_datetime(df0['departure_time']).dt.hour * 60

df0['arrival_time'] = pd.to_datetime(df0['arrival_time']).dt.minute + \
                      pd.to_datetime(df0['arrival_time']).dt.hour * 60
```

```python
# Time interval
df0['travel_time'] = df0['travel_time'].apply(lambda x: pd.to_timedelta(x).total_seconds() / 60)
```

```python
df0
```

### Train/Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
y,X = df0['price'], df0.drop(columns=['price'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

### Polynomial Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Fit a polynomial regression model to the training data
poly_reg = PolynomialFeatures(degree=4)
X_train_poly = poly_reg.fit_transform(X_train)
lin_reg = LinearRegression()
lin_reg.fit(X_train_poly, y_train)
```

```python
# Evaluate the model on the testing data
X_test_poly = poly_reg.transform(X_test)
y_pred = lin_reg.predict(X_test_poly)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Compute predictions and residuals
residuals = y_test - y_pred

# Compute MSE
mse = np.mean(residuals**2)

# Visualize residuals
plt.scatter(y_pred, residuals)
plt.xlabel("Predictions")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```
